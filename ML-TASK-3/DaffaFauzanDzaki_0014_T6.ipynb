# Classification Task dengan Decision Tree

## Pengenalan komponen Decision Tree: root, node, leaf

Biru : Root Node
Merah : Internal Node
Kuning : Leaf Node

## Gini Impurity

### Ruas Kiri

$G=1-\sum_{i}^n P_i^2$

$\quad = 1 - P(biru)^2$

$\quad =1-(\frac{4}{4})^2=0$

### Ruas Kanan

$G=\sum_{i}^n P_i^2$

$\quad=1-(P(biru)^2 + P(hijau)^2)$

$\quad=1-((\frac{1}{6})^2 + (\frac{5}{6})^2)=0.278$

### Average Gini Impurity

$G = \frac{4}{4+6}\times0+\frac{6}{4+6}\times0.278$

$\quad = 0.1668$

## Membangun Decision Tree

$\quad=1-((\frac{2}{5})^2+(\frac{2}{5})^2+(\frac{1}{5})^2)$

$\quad=0.63$

## Dataset

```python
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)

print(f'Dimensi Feature: {X.shape}')
print(f'Class: {set(y)}')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,
y,
test_size=0.3,
random_state=0)
```

## Classification dengan DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth=4)

model.fit(X_train, y_train)
```

## Visualisasi Model

```python
import matplotlib.pyplot as plt
from sklearn import tree

plt.rcParams ['figure.dpi'] = 85
plt.subplots(figsize=(10, 10))
tree.plot_tree(model, fontsize=10)
plt.show()
```

## Evaluasi Model

```python
from sklearn.metrics import classification_report

y_pred = model. predict(X_test)

print(classification_report(y_test, y_pred))
```
